**Notebook to illustrate the existence of non-compact invariant manifolds that are locally completely unstable and yet, are globally stable.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

**We start by constructing a dynamical system that possesses a non-compact invariant manifold that is locally completely unstable and yet, is globally stable.**

```python
# Defining a dynamical system with the required behavior
def dynamic_system1(init_pos, t, l1, l2, omega, k):
    
    x = init_pos[0]
    
    y = init_pos[1]
    
    z = init_pos[2]
    
    R = np.matrix([[np.cos(omega*z), -np.sin(omega*z)], [np.sin(omega*z), np.cos(omega*z)]])
    
    A = R@np.matrix([[l1, 0],[0, l2]])@np.linalg.inv(R)
    
    dx_dt = A[0,0]*x+A[0,1]*y
    
    dy_dt = A[1,0]*x+A[1,1]*y

    dz_dt = k*(1+np.sin(z)**2)
    
    return dx_dt,dy_dt,dz_dt
```

```python
# Choosing parameters of the dynamical system that causes it to exhibit interesting dynamics

# Eigenvalues controlling x-y dynamics:

# EV1<0
l1 = -0.5
# EV2>0
l2 = 0.2

# Speed along the z-axis
k = 0.01
```

**In the example system defined above, the z-axis acts as an invariant manifold, while the dynamics normal to the z-axis is determined by the 'x' and 'y' directional dynamics. The eigenvalues $\lambda_1$ and $\lambda_2$ are chosen such that the z-axis is unstable or repelling at every point. However, we shall see that the causing the stable and unstable eigenvectors associated with $\lambda_1$ and $\lambda_2$ to rotate above a critical rate of rotation $\omega$ causes the z-axis to become attracting to the trajectories instead.**

**Now we numerically investigate qualitatively different cases of dynamics occuring for various values of $\omega$.
See descriptions below images for explanations of various cases.**

```python
# Case 1 with \omega=0
omega = 0
#Integration time
t = np.linspace(1, 30, 300)

#Integrating 4 different initial conditions
init_pos1 = [-.1, .1, .1]

pos1= odeint(dynamic_system1, init_pos1, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)

init_pos2 = [-.1, .1, .2]

pos2= odeint(dynamic_system1, init_pos2, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)

init_pos3 = [-.1, .1, .3]

pos3= odeint(dynamic_system1, init_pos3, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)

init_pos4 = [-.1, .1, .4]

pos4= odeint(dynamic_system1, init_pos4, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)
```

```python

#Plotting the trajectories

#The initial phase-space coordinates are indicated as cyan bubbles while the final coordinates are indicated in red
z_axis = np.linspace(.1,.5,100)
ax = plt.figure().add_subplot(projection='3d')
ax.plot(np.zeros(100),np.zeros(100),z_axis, c ='k', linewidth =5)
ax.plot(pos1[:,0],pos1[:,1],pos1[:,2],c='b')
ax.plot(pos2[:,0],pos2[:,1],pos2[:,2],c='r')
ax.plot(pos3[:,0],pos3[:,1],pos3[:,2],c='g')
ax.plot(pos4[:,0],pos4[:,1],pos4[:,2],c='y')

ax.scatter(pos1[0,0],pos1[0,1],pos1[0,2],c='c')
ax.scatter(pos2[0,0],pos2[0,1],pos2[0,2],c='c')
ax.scatter(pos3[0,0],pos3[0,1],pos3[0,2],c='c')
ax.scatter(pos4[0,0],pos4[0,1],pos4[0,2],c='c')

ax.scatter(pos1[-1,0],pos1[-1,1],pos1[-1,2],c='r')
ax.scatter(pos2[-1,0],pos2[-1,1],pos2[-1,2],c='r')
ax.scatter(pos3[-1,0],pos3[-1,1],pos3[-1,2],c='r')
ax.scatter(pos4[-1,0],pos4[-1,1],pos4[-1,2],c='r')
```

**Case 1: $\omega=0$, this causes the z-axis to behave merely as a line of saddle type fixed points with trajectories approaching the z-axis briefly before diverging away from it as t-->$\infty$**

```python
# Case 2 with \omega=10
omega = 10
#Integration time
t = np.linspace(1, 30, 300)

#Integrating 4 different initial conditions
init_pos1 = [-.1, .1, .1]

pos1= odeint(dynamic_system1, init_pos1, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)

init_pos2 = [-.1, .1, .2]

pos2= odeint(dynamic_system1, init_pos2, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)

init_pos3 = [-.1, .1, .3]

pos3= odeint(dynamic_system1, init_pos3, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)

init_pos4 = [-.1, .1, .4]

pos4= odeint(dynamic_system1, init_pos4, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)
```

```python
#Plotting the trajectories

#The initial phase-space coordinates are indicated as cyan bubbles while the final coordinates are indicated in redz_axis = np.linspace(.1,.5,100)
ax = plt.figure().add_subplot(projection='3d')
z_axis = np.linspace(.1,.5,100)
ax.plot(np.zeros(100),np.zeros(100),z_axis, c ='k', linewidth =5)
ax.plot(pos1[:,0],pos1[:,1],pos1[:,2],c='b')
ax.plot(pos2[:,0],pos2[:,1],pos2[:,2],c='r')
ax.plot(pos3[:,0],pos3[:,1],pos3[:,2],c='g')
ax.plot(pos4[:,0],pos4[:,1],pos4[:,2],c='y')

ax.scatter(pos1[0,0],pos1[0,1],pos1[0,2],c='c')
ax.scatter(pos2[0,0],pos2[0,1],pos2[0,2],c='c')
ax.scatter(pos3[0,0],pos3[0,1],pos3[0,2],c='c')
ax.scatter(pos4[0,0],pos4[0,1],pos4[0,2],c='c')

ax.scatter(pos1[-1,0],pos1[-1,1],pos1[-1,2],c='r')
ax.scatter(pos2[-1,0],pos2[-1,1],pos2[-1,2],c='r')
ax.scatter(pos3[-1,0],pos3[-1,1],pos3[-1,2],c='r')
ax.scatter(pos4[-1,0],pos4[-1,1],pos4[-1,2],c='r')

ax.view_init(elev=11, azim=230)
```

**Case 2: $\omega=10$, Here, the rate of rotation is insufficient to significantly alter the dynamics from case 1, with the trajectories diverging away from the z-axis as t-->$\infty$. Note that while the eventual fate of the trajectories are qualitatively similar, their geometry is influenced by the rotation of the eigenvectors.**

```python
# Case 3 with \omega=50
omega = 50
#Integration time
t = np.linspace(1, 30, 300)

#Integrating 4 different initial conditions
init_pos1 = [-.1, .1, .1]

pos1= odeint(dynamic_system1, init_pos1, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)

init_pos2 = [-.1, .1, .2]

pos2= odeint(dynamic_system1, init_pos2, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)

init_pos3 = [-.1, .1, .3]

pos3= odeint(dynamic_system1, init_pos3, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)

init_pos4 = [-.1, .1, .4]

pos4= odeint(dynamic_system1, init_pos4, t, args=(l1, l2, omega, k), rtol=10**-10, atol=10**-10)
```

```python
#Plotting the trajectories

#The initial phase-space coordinates are indicated as cyan bubbles while the final coordinates are indicated in redz_axis = np.linspace(.1,1,100)
ax = plt.figure().add_subplot(projection='3d')
z_axis = np.linspace(.1,1,100)
ax.plot(np.zeros(100),np.zeros(100),z_axis, c ='k', linewidth =5)
ax.plot(pos1[:,0],pos1[:,1],pos1[:,2],c='b')
ax.plot(pos2[:,0],pos2[:,1],pos2[:,2],c='r')
ax.plot(pos3[:,0],pos3[:,1],pos3[:,2],c='g')
ax.plot(pos4[:,0],pos4[:,1],pos4[:,2],c='y')

ax.scatter(pos1[0,0],pos1[0,1],pos1[0,2],c='c')
ax.scatter(pos2[0,0],pos2[0,1],pos2[0,2],c='c')
ax.scatter(pos3[0,0],pos3[0,1],pos3[0,2],c='c')
ax.scatter(pos4[0,0],pos4[0,1],pos4[0,2],c='c')

ax.scatter(pos1[-1,0],pos1[-1,1],pos1[-1,2],c='r')
ax.scatter(pos2[-1,0],pos2[-1,1],pos2[-1,2],c='r')
ax.scatter(pos3[-1,0],pos3[-1,1],pos3[-1,2],c='r')
ax.scatter(pos4[-1,0],pos4[-1,1],pos4[-1,2],c='r')
```

**Case 3: $\omega=50$, While the z-axis still behaves as a line of saddle type unstable equilibrium points in their immediate vicinity, the asymptotic behavior of the trajectories is to approach the z-axis as t-->$\infty$ thus making the z-axis globally stable despite the local complete instability of the z-axis.**
